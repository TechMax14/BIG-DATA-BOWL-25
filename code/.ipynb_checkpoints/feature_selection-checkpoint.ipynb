```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go 
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
from IPython.display import display
import time
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor


# Load the datasets (ensure correct paths and files)
train_features = pd.read_csv('../data/processed_data/merged_train_data.csv')
test_features = pd.read_csv('../data/processed_data/week_9_test_features.csv')

#train_labels = pd.read_csv('../data/processed_data/merged_train_target.csv')
```

```python
# DROP UNNECESSARY COLUMNS
print(train_features.columns)

# Define the columns to drop
columns_to_drop = ['gameId', 'playId', 'isDropback', 'week', 'yardsGained', 'receiverAlignment']

# Drop only the columns that exist in train_features
train_features = train_features.drop(columns=[col for col in columns_to_drop if col in train_features.columns])

# # Create a mapping for receiverAlignment
# alignment_mapping = {
#     '0x0': 0,
#     '1x0': 1,
#     '1x1': 2,
#     '2x0': 3,
#     '2x1': 4,
#     '2x2': 5,
#     '3x0': 6,
#     '3x1': 7,
#     '3x2': 8,
#     '3x3': 9,
#     '4x1': 10,
#     '4x2':11
# }

# # Apply the mapping
# train_features['receiverAlignment'] = train_features['receiverAlignment'].map(alignment_mapping)

print(train_features.columns)
```

```python
# Remove 4 rows missing values
train_features = train_features.dropna(subset=['offenseFormation'])
# Remove 4 rows missing values
train_features = train_features.dropna(subset=['pre_snap_time_duration'])

# Assuming train_features contains all your features and 'pff_manZone' is the column you want to predict
train_labels = train_features['play_type']

# Drop 'pff_manZone' from train_features to avoid data leakage
train_features = train_features.drop(columns=['play_type'])

target_corr = train_features.corrwith(train_labels)  # Assuming `train_labels` contains your target variables
print(target_corr)
```

```python
# List of features to remove due to low correlation
features_to_remove = [
    'absoluteYardlineNumber',
    'quarter',
    'motion_player_count'
]

train_features = train_features.drop(columns=features_to_remove)
test_features = test_features.drop(columns=features_to_remove)

# Calculate VIF for each feature
X = train_features # Without the target
#X = X.dropna(subset=['pre_snap_time_duration']) # Removing 4 NA rows
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

```python
# List of features to remove due to low correlation
features_to_remove = [
    'expectedPoints',
    'expectedPointsAdded',
    'field_position_weight',
    'playSuccessWeight',
    'possessionTeamWinProbability',
    'possessionTeamImpact',
    'opponentTeamImpact',
    'players_in_box_count',
]

train_features = train_features.drop(columns=features_to_remove)
test_features = test_features.drop(columns=features_to_remove)

# Calculate VIF for each feature
X = train_features # Without the target
#X = X.dropna(subset=['pre_snap_time_duration']) # Removing 4 NA rows
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```
